Computing deviation of the evaluator scores

```python


def std_dev(df):
    scores=np.array(df['scores'])
    scores_std=scores.std()
    scores_mean=scores.mean()
    df['scores_std']=scores_std
    df['scores_mean']=scores_mean
    
    return df
```

```python
df=df_existing_oup.copy()
df['scores']=df['scores'].apply(lambda x: ast.literal_eval(x))
df.loc[:, ['scr1', 'scr2', 'scr3', 'scr4']]=df["scores"].to_list()
df=df.apply(std_dev, axis=1)
df_existing_oup=df.copy()
df_existing_oup
```

```python
df=df_minnie_oup.copy()
df['scores']=df['scores'].apply(lambda x: ast.literal_eval(x))
df.loc[:, ['scr1', 'scr2', 'scr3', 'scr4']]=df["scores"].to_list()
df=df.apply(std_dev, axis=1)
df_minnie_oup=df.copy()
df_minnie_oup
```

```python
df=df_singleprompt_oup.copy()
df['scores']=df['scores'].apply(lambda x: ast.literal_eval(x))
df.loc[:, ['scr1', 'scr2', 'scr3', 'scr4']]=df["scores"].to_list()
df=df.apply(std_dev, axis=1)
df_singleprompt_oup=df.copy()
df_singleprompt_oup
```

```python


def coefficients(x, y):
    pearson=stats.pearsonr(x, y)
    spearmanr=stats.spearmanr(x,y)
    kendalltau=stats.kendalltau(x,y)
    
    return pearson, spearmanr, kendalltau
```

```python
#Existing_oup
x=df_existing_oup['mean'].to_numpy()
y=df_existing_oup['second_prompt_output'].to_numpy()
coefficients(x,y)
```

```python
#Existing_oup
x=df_minnie_oup['mean'].to_numpy()
y=df_minnie_oup['second_prompt_output'].to_numpy()
coefficients(x,y)
```

```python
#Existing_oup
x=df_singleprompt_oup['mean'].to_numpy()
y=df_singleprompt_oup['first_prompt_output'].to_numpy()
coefficients(x,y)
```

Computing coefficient (annotator output & Prompt output) for a range of standard deviation of the annotator scores 

```python
std_dev_bucket=[4,5,6,7,8,9,10,15,100]


for sd in std_dev_bucket:
    df=df_minnie_oup.loc[(df_minnie_oup['scores_std']>3) & (df_minnie_oup['scores_std']<sd)]
    x=df['mean'].to_numpy()
    y=df['second_prompt_output'].to_numpy()
    s, p, k=coefficients(x,y)
    print(' std_dev less than ',sd, '\n','num_elements ', len(df), '\n','spearman coeff ', s[0] , '\n')
    
```
